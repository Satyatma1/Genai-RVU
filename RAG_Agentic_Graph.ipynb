Agentic RAG Implementation
Concept: An Agentic RAG doesn't just search; it decides. If the user asks a question that requires a calculation or a specific tool, the agent chooses the right tool instead of just pulling text.

```python
# Simple Simulation of an Agentic RAG Flow
class AgenticRAG:
    def __init__(self):
        self.knowledge_base = {
            "satyatma": "Satyatma Chincholi's USN is 1RVU23CSE414.",
            "lab": "The AI Lab is in Room 402."
        }

    def search_tool(self, query):
        print(f"[Tool: Search] Searching for '{query}'...")
        for key in self.knowledge_base:
            if key in query.lower():
                return self.knowledge_base[key]
        return "No data found."

    def calculator_tool(self, expression):
        print(f"[Tool: Calculator] Solving {expression}...")
        try: return eval(expression)
        except: return "Error"

    def run_agent(self, user_prompt):
        # The 'Brain' (LLM) decides which tool to use
        print(f"User Question: {user_prompt}")

        if "usn" in user_prompt.lower():
            action = "search"
            result = self.search_tool("satyatma")
        elif "+" in user_prompt or "*" in user_prompt:
            action = "calculate"
            result = self.calculator_tool(user_prompt.split("is")[-1].strip())
        else:
            return "I'm not sure how to help with that."

        return f"Agent Conclusion: {result}"

# Test
agent = AgenticRAG()
print(agent.run_agent("What is the USN for Satyatma?"))
print("-" * 30)
print(agent.run_agent("What is 1024 * 2?"))
```

Graph RAG Implementation
Concept: Instead of just looking for keywords, Graph RAG looks at the relationships between entities (Nodes and Edges). This is powerful for answering questions like "Who is the student in Room 402?" by traversing links.

```python
import networkx as nx

# 1. Create the Knowledge Graph
G = nx.Graph()

# Adding entities (Nodes) and relationships (Edges)
G.add_edge("Satyatma Chincholi", "1RVU23CSE414", relation="has_usn")
G.add_edge("Satyatma Chincholi", "Deep Learning Lab", relation="attends")
G.add_edge("Deep Learning Lab", "Room 402", relation="located_in")

def graph_rag_query(entity, goal_relation):
    print(f"Querying Graph for: {entity} -> {goal_relation}")

    # Simple traversal: Find neighbors of the entity
    if entity in G:
        for neighbor in G.neighbors(entity):
            edge_data = G.get_edge_data(entity, neighbor)
            if edge_data['relation'] == goal_relation:
                return f"Found: {neighbor}"

        # If not found directly, try a 2-step hop (Multi-hop RAG)
        for neighbor in G.neighbors(entity):
            for second_neighbor in G.neighbors(neighbor):
                edge_data = G.get_edge_data(neighbor, second_neighbor)
                if edge_data['relation'] == goal_relation:
                    return f"Multi-hop Discovery: {entity} is connected to {second_neighbor} via {neighbor}"

    return "Relationship not found in graph."

# Test
print(graph_rag_query("Satyatma Chincholi", "has_usn"))
print("-" * 30)
# Find where Satyatma's lab is located (2 hops away)
print(graph_rag_query("Satyatma Chincholi", "located_in"))
```
